```python
import pandas as pd
from sqlalchemy import create_engine, text
```

```python
police_kpi = pd.read_csv('../data/KPI_Response_Full_Data_data.csv')
```

```python
database_name = 'Metro_Police_Response_Times'    

connection_string = f"postgresql://postgres:postgres@localhost:5432/{database_name}"
```

```python
engine = create_engine(connection_string)
```

```python
police_response_redux = 'Select zip, precinct, call_rec, tencode_desc, response_code, received_to_dispatch, dispatch_to_arrival, received_to_arrival, received_to_complete, one From police_kpi'
```

```python
with engine.connect() as connection:
    police_response_redux = pd.read_sql(text(police_response_redux), con = connection)

police_response_redux.head()
```

```python
median_income = pd.read_csv('../data/census_median_income.csv')
```

```python
median_income['location'] = median_income['location'].astype(int)
police_response_vs_income_redux = pd.merge(median_income, police_response_redux, left_on='location', right_on='zip',how='inner')
```

```python
police_response_vs_income_redux = police_response_vs_income_redux.drop(columns=['zip'])
police_response_vs_income_redux = police_response_vs_income_redux.rename(columns={'location': 'zipcode'})
```

```python
police_response_vs_income_redux
```

```python
police_response_vs_income_redux.to_csv('police_response_vs_income_final.csv', index=False)
```
