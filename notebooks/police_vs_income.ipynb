```python
import pandas as pd
```

```python
median_income = pd.read_csv('../data/census_median_income.csv')
police_kpi = pd.read_csv('../data/police_kpi_final.csv')
```

```python
median_income['location'] = median_income['location'].astype(int)
police_vs_income_final = pd.merge(median_income, police_kpi, left_on='location', right_on='zip',how='inner')
```

```python
police_vs_income_final = police_vs_income_final.drop(columns=['zip'])
police_vs_income_final = police_vs_income_final.rename(columns={'location': 'zipcode'})
```

```python
police_vs_income_37215 = police_vs_income_final[police_vs_income_final['zipcode'] == 37215]
police_vs_income_37215
```

```python
police_vs_income_final.to_csv('police_vs_income_final.csv', index=False)
```
